#Q1. What is a Probability Density Function?
The probability density function (pdf) is a function describing the probability of a continuous random variable taking on a particular value. The pdf does not give the probability that the variable has some particular value-the probability that the variable takes on any one particular value is always zero for continuous distributions-but gives the relative probability that the variable is close to a particular value.

The space underneath the PDF curve during a period determines the probability that the random variable is in the space interval. The overall space below the curve equals 1.

#Q2: Types of Probability Distributions?
There are two broad types of categories of probability distributions,

Discrete Probability Distributions:
Examples, Binomial, Poisson, Geometric.
Random variable countable number values.

Continuous Probability Distributions:
Examples, Normal, Uniform, Exponential.
Applied for the random variables which take infinite values in an interval.

# Q3: Python Function for the PDF of a Normal Distribution
import math
def normal_pdf(x, mean, std_dev):
    """
    Computes the probability density function of a normal distribution.
    
    Parameters:
        x (float): evaluation point of the PDF.
        mean (float): mean of the distribution.
        std_dev (float): standard deviation of the distribution.
        
    Returns:
        float: PDF value at x.
    """
    coefficient = 1 / (std_dev * math.sqrt(2 * math.pi))
    exponent = math.exp(-0.5 * ((x - mean) / std_dev) ** 2)
    return coefficient * exponent

#Q4: Properties of Binomial Distribution
A fixed number of trials (n): The experiment is carried out a number of times.
Two outcomes: Each trial is either success or failure.
Independent trials: The result of any one trial is not influenced by another.
Same probability (p): The same probability of success for any trial
Example:
Flipping a coin 10 times and observing the number of heads obtained.
Testing 20 bulbs to observe how many of them are defective.

#Q6: Python Function for CDF of Poisson Distribution
from math import exp, factorial
"""

def poisson_cdf(k, mean):
    """
    Calculate the cumulative distribution function of a Poisson distribution.
    
    Parameters:
        k (int): Point at which to evaluate the CDF.
         mean (float): Mean of the Poisson distribution.
    Returns
    -------
    float: CDF value at k.
    """
    cdf = sum((mean**i * exp(-mean)) / factorial(i) for i in range(0, k+1))
    return cdf

#Q7: What is the Difference Between Binomial Distribution and Poisson Distribution?
Nature of Trials:

Binomial: Number of trials is fixed.
Poisson: Number of events in a fixed interval or space.
Parameters:

Binomial: Requires number of trials (n) and probability of success (p).
Poisson: Defined by the mean (λ).
Type of Outcome:

Binomial: Discrete outcome based on successes in trial.
Poisson: Discrete outcome based on the count of events during an interval.

#Q10: In Normal Distribution, Where Does the Least Frequent Data Appear?
In a normal distribution, the least frequent data occurs at the extremes of the distribution, or away from the mean. It occurs at the tails of the bell curve because the probability density drops quickly as you get away from the mean.
