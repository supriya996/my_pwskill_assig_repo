Q1. What are the three measures of central tendency?

Answer: The three measures of central tendency are mean, median, and mode.

***

Q2. What is the difference between the mean, median, and mode? How are they used to measure the central tendency of a dataset?

Answer:

Mean: It is the average of all values in the dataset. It is calculated by summing all the values and dividing by the total number of values. Sensitive to outliers.

Median: The middle value in a sorted dataset. In case of an even number of values, it is the average of the two middle values. Less affected by outliers.

Mode: The value that occurs most frequently in the dataset. A dataset can have one mode, multiple modes, or no mode. They are used to describe where the center or middle of the dataset is located but can be affected by outliers differently.

Q5. How do measures of dispersion like range, variance, and standard deviation describe the spread of a dataset? Give an example.



Answer: Measures of dispersion describe the spread or variability of data points around the central tendency.

Range: It is the difference between the maximum and minimum values in a dataset. It portrays the total spread but is sensitive to outliers.
Variance: It is the average of the squared differences from the mean. It gives an idea of how far the data points are spread from the mean.
Standard Deviation: It is the square root of the variance. It is a commonly used measure of spread because it is in the same units as the original data. Example: Take two sets A = [10, 20, 30, 40, 50] and B = [30, 30, 30, 30, 30]. The average of both sets is 30. Range: For set A is 50-10 = 40. For set B is 30-30=0. Variance : For A is 200. For B is 0. Standard deviation: For A is approximately 14.14. For B is 0. This means that the spread of the dataset A is much wider than B, which has no dispersion.

Q8. What is your understanding about skewness of data?


Answer: Skewness is a measure of the asymmetry of a probability distribution. It indicates whether the distribution is symmetrical or if it leans towards one side. Positive skewness (right-skewed) indicates a longer tail on the right side, and negative skewness (left-skewed) indicates a longer tail on the left side.

***
Q9. If a data is right skewed then what will be the position of median with respect to mean?


Answer: In a right-skewed distribution, the mean is usually larger than the median because the mean is pulled to the longer right tail by the large values. Therefore, the median will be to the left of the mean.

Q10. What is the difference between covariance and correlation? How are these measures used in statistical analysis?

***

Answer:

Covariance: Measures how two variables change together. Positive covariance indicates that both the variables increase together, while the negative covariance indicates that when one increases, the other decreases. However, it is not scaled and hence gives no idea of the strength of relation and cannot be compared to that with covariance.
Correlation: Measures the strength as well as the direction of a linear relationship between two variables. It is a standardized measure ranging from -1 to 1. A correlation of 1 indicates perfect positive linear relationship, -1 indicates perfect negative linear relationship, and 0 indicates no linear relationship. Use in Statistical Analysis: Covariance provides a basic measure of dependence, but correlation is more useful for determining how strongly variables are related. Correlation is used in regression analysis and to understand the strength of relationships.

Q11. What is the formula for the sample mean? Provide an example of a calculation on a set of data.


Answer: The formula to find the sample mean (x̄) is as follows: x̄ = Σxᵢ/n, where the summation of all the values in the sample is Σxᵢ and n is the number of samples. Example: Sample data: [10, 12, 15, 18, 20] Sample mean = (10 + 12 + 15 + 18 + 20) / 5 = 75/5 = 15.

Answer: In case of normal distribution, mean, median and mode are all equal and they are all at the center.

Q13. Covariance and correlation differ from each other in what aspect?


Answer: Covariance shows how two variables change together, but its value does not have a set limit and is not standardized and thus, does not specify the strength of the relationship. Correlation is a standardized measure (between -1 and 1) that shows the strength and direction of the linear relationship.

Q14. How do outliers affect measures of central tendency and dispersion? Give an example.

***
Answer:

Outliers significantly affect the mean because the mean takes all values into consideration and gets pulled in the direction of the extreme values, outliers.
The median is less affected by outliers as it is based on position, not the extreme values.
Outliers greatly increase measures of dispersion such as range, variance, and standard deviation as the deviations are calculated in reference to the mean, which is shifted. Example: Take the data set [10, 12, 15, 18, 20]. Mean = 15, median =15. Now take the outlier 100 is included: [10, 12, 15, 18, 20, 100]. The mean now becomes 29.16 and median becomes 16.5. The outlier has pulled the mean much more than it pulled the median and it has significantly widened the spread of the data set.
